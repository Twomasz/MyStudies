# Zadanie domowe

W przypadku obrazów w odcieniach szarości pojedynczy piksel z zakresu [0; 255] reprezentowany jest jako 8-bitowa liczba bez znaku.
Pewnym rozszerzeniem analizy sposobu reprezentacji obrazu może być następujący eksperyment.
Załóżmy, że z każdego z 8 bitów możemy stworzyć pojedynczy obraz binarny (ang. _bit-plane slicing_).
Dla obrazka _100zloty.jpg_ (https://raw.githubusercontent.com/vision-agh/poc_sw/master/02_Point/100zloty.jpg) stwórz 8 obrazów, z których każdy powinien zawierać jedną płaszczyznę bitową.
Podpowiedź $-$ warto sprawdzić, jak realizuje się bitowe operacje logiczne.
Zastosowanie takiej dekompozycji obrazu pozwala na analizę ,,ważności'' poszczególnych bitów.
Jest to użyteczne w kwantyzacji, ale także w kompresji.

W drugim etapie zadania proszę spróbować odtworzyć obraz oryginalny z mniejszej liczby obrazów binarnych.
Warto zacząć od dwóch najbardziej znaczących bitów, a później dodawać kolejne.
Należy utworzyć co najmniej trzy wersje zrekonstruowanych obrazów.
Podpowiedź $-$ rekonstrukcja obrazu to mnożenie przez odpowiednią potęgę liczby 2 (przesunięcie bitowe) oraz dodawanie.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os

if not os.path.exists('100zloty.jpg'):
    !wget https://raw.githubusercontent.com/vision-agh/poc_sw/master/02_Point/100zloty.jpg
```

```python
cash = cv2.imread('100zloty.jpg')
gray_cash = cv2.cvtColor(cash, cv2.COLOR_RGB2GRAY)
plt.imshow(gray_cash)
plt.gray()
plt.show()
```

```python
row_size, col_size = len(gray_cash), len(gray_cash[0])

first_bit = [[] for x in range(row_size)]
for i in range(row_size):
    for j in range(col_size):
        first_bit[i].append(gray_cash[i][j] & 0b10000000)
        
plt.figure(figsize=(8, 8))
plt.imshow(first_bit)
plt.title('Najstarszy bit')
plt.show()

second_bit = [[] for x in range(row_size)]
for i in range(row_size):
    for j in range(col_size):
        second_bit[i].append(gray_cash[i][j] & 0b1000000)

plt.figure(figsize=(8, 8))
plt.imshow(second_bit)
plt.title('Drugi bit')
plt.show()

third_bit = [[] for x in range(row_size)]
for i in range(row_size):
    for j in range(col_size):
        third_bit[i].append(gray_cash[i][j] & 0b100000)

plt.figure(figsize=(8, 8))
plt.imshow(third_bit)
plt.title('Trzeci bit')
plt.show()

fourth_bit = [[] for x in range(row_size)]
for i in range(row_size):
    for j in range(col_size):
        fourth_bit[i].append(gray_cash[i][j] & 0b10000)

plt.figure(figsize=(8, 8))
plt.imshow(fourth_bit)
plt.title('Czwarty bit')
plt.show()

fifth_bit = [[] for x in range(row_size)]
for i in range(row_size):
    for j in range(col_size):
        fifth_bit[i].append(gray_cash[i][j] & 0b1000)

plt.figure(figsize=(8, 8))
plt.imshow(fifth_bit)
plt.title('Piąty bit')
plt.show()

sixth_bit = [[] for x in range(row_size)]
for i in range(row_size):
    for j in range(col_size):
        sixth_bit[i].append(gray_cash[i][j] & 0b100)

plt.figure(figsize=(8, 8))
plt.imshow(sixth_bit)
plt.title('Szósty bit')
plt.show()

seventh_bit = [[] for x in range(row_size)]
for i in range(row_size):
    for j in range(col_size):
        seventh_bit[i].append(gray_cash[i][j] & 0b10)

plt.figure(figsize=(8, 8))
plt.imshow(seventh_bit)
plt.title('Siódmy bit')
plt.show()

youngest_bit = [[] for x in range(row_size)]
for i in range(row_size):
    for j in range(col_size):
        youngest_bit[i].append(gray_cash[i][j] & 0b1)

plt.figure(figsize=(8, 8))
plt.imshow(youngest_bit)
plt.title('Najmłodszy bit')
plt.show()
```

### Komentarz:
Najbardziej znaczącą warstwą okazała się ta zawierająca najstarsze bity, gdyż przedstawia ona wartości odcienia szarości z przedziału **(0, 127)** jako **czarny**, a wartości z przedziału **(128, 255)** jako **biały**. Natomiast im bliżej najmłodszego bitu, tym mniej czytelny staje się obraz, bo tak jak w przypadku wzięcia pod uwagę ostatniego bitu z dwóch wartości np. **244** i  **245**, dają dwa skrajne odcienie, mimo że wejściowo mają tak podobny do siebie odcień, że widząc je, nie będziemy w stanie dostrzec różnicy. Powoduje to powstanie charakterystycznego śnieżącego obrazu, który na pierwszy rzut oka nie zawiera żadnych informacji.
