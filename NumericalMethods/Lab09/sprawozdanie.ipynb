***Zadanie 6.***  
Dla głównej funkcji oraz przedziału zdefiniowanego z zadania 1 znajdź miejsca zerowe przy użyciu:

* metody bisekcji
* metody siecznych
* metody Newtona

z tolerancją $10^{-10}$

Zbadaj dokładność(względem rozwiązania z zadania 5) i czas obliczeń metod w zależności od liczby iteracji. Wyniki przedstaw na wykresach. 

```python
# Dane do funkcji

a1 = -1.0
b1 = 0.2

a2 = 0.5
b2 = 3.0

epsilon = 10e-10
it = [1, 3, 5, 10, 15, 20, 25, 30, 35, 40, 45, 50, 55, 60]
```

```python
# METODA BIJEKCJI

bisection_norm_ans1 = []
bisection_norm_ans2 = []

for i in range(len(it)):
    ans1 = main.bisection(a1, b1, fun, epsilon, it[i])
    ans2 = main.bisection(a2, b2, fun, epsilon, it[i])
    bisection_norm_ans1.append(np.abs(real_ans[0] - ans1[0]))
    bisection_norm_ans2.append(np.abs(real_ans[1] - ans2[0]))
    
plt.scatter(it, bisection_norm_ans1)
plt.grid()
plt.title('Rozwiązanie 1-go miejsca zerowego metodą bisekcji')
plt.xlabel('Liczba iteracji')
plt.ylabel('Dokładność rozwiązania')
plt.semilogy()
plt.show()

time = []
error = []
for i in range(len(it)):
    method_time = %timeit -r 10 -n 10 -o main.bisection(a1, b1, fun, epsilon, it[i])
    time.append(method_time.average)
    error.append(method_time.stdev)
    plt.errorbar(it[i], time[i], yerr=error[i], marker="x")
    plt.title('Zależnośc czasu obliczania od maksymalnej liczby iteracji')
    plt.xlabel('Liczba iteracji')
    plt.ylabel('Czas obliczeń')
    plt.grid()
plt.show()

plt.scatter(it, bisection_norm_ans2)
plt.grid()
plt.title('Rozwiązanie 2-go miejsca zerowego metodą bisekcji')
plt.xlabel('Liczba iteracji')
plt.ylabel('Dokładność rozwiązania')
plt.semilogy()
plt.show()

time = []
error = []
for i in range(len(it)):
    method_time = %timeit -r 10 -n 10 -o main.bisection(a2, b2, fun, epsilon, it[i])
    time.append(method_time.average)
    error.append(method_time.stdev)
    plt.errorbar(it[i], time[i], yerr=error[i], marker="x")
    plt.title('Zależnośc czasu obliczania od maksymalnej liczby iteracji')
    plt.xlabel('Liczba iteracji')
    plt.ylabel('Czas obliczeń')
    plt.grid()
plt.show()
```

```python
# METODA SIECZNEJ

secant_norm_ans1 = []
secant_norm_ans2 = []

for i in range(len(it)):
    ans1 = main.secant(a1, b1, fun, epsilon, it[i])
    ans2 = main.secant(a2, b2, fun, epsilon, it[i])
    secant_norm_ans1.append(np.abs(real_ans[0] - ans1[0]))
    secant_norm_ans2.append(np.abs(real_ans[1] - ans2[0]))
    
plt.scatter(it, secant_norm_ans1)
plt.grid()
plt.title('Rozwiązanie 1-go miejsca zerowego metodą siecznej')
plt.xlabel('Liczba iteracji')
plt.ylabel('Dokładność rozwiązania')
plt.semilogy()
plt.show()

time = []
error = []
for i in range(len(it)):
    method_time = %timeit -r 10 -n 10 -o main.secant(a1, b1, fun, epsilon, it[i])
    time.append(method_time.average)
    error.append(method_time.stdev)
    plt.errorbar(it[i], time[i], yerr=error[i], marker="x")
    plt.title('Zależnośc czasu obliczania od maksymalnej liczby iteracji')
    plt.xlabel('Liczba iteracji')
    plt.ylabel('Czas obliczeń')
    plt.grid()
plt.show()

plt.scatter(it, secant_norm_ans2)
plt.grid()
plt.title('Rozwiązanie 2-go miejsca zerowego metodą siecznej')
plt.xlabel('Liczba iteracji')
plt.ylabel('Dokładność rozwiązania')
plt.semilogy()
plt.show()

time = []
error = []
for i in range(len(it)):
    method_time = %timeit -r 10 -n 10 -o main.secant(a2, b2, fun, epsilon, it[i])
    time.append(method_time.average)
    error.append(method_time.stdev)
    plt.errorbar(it[i], time[i], yerr=error[i], marker="x")
    plt.title('Zależnośc czasu obliczania od maksymalnej liczby iteracji')
    plt.xlabel('Liczba iteracji')
    plt.ylabel('Czas obliczeń')
    plt.grid()
plt.show()
```

```python
# METODA NEWTONA

newton_norm_ans1 = []
newton_norm_ans2 = []

for i in range(len(it)):
    ans1 = main.newton(fun, dfun, ddfun, a1, b1, epsilon, it[i])
    ans2 = main.newton(fun, dfun, ddfun, a2, b2, epsilon, it[i])
    newton_norm_ans1.append(np.abs(real_ans[0] - ans1[0]))
    newton_norm_ans2.append(np.abs(real_ans[1] - ans2[0]))
    
plt.scatter(it, newton_norm_ans1)
plt.grid()
plt.title('Rozwiązanie 1-go miejsca zerowego metodą Newtona')
plt.xlabel('Liczba iteracji')
plt.ylabel('Dokładność rozwiązania')
plt.semilogy()
plt.show()

time = []
error = []
for i in range(len(it)):
    method_time = %timeit -r 10 -n 10 -o main.newton(fun, dfun, ddfun, a1, b1, epsilon, it[i])
    time.append(method_time.average)
    error.append(method_time.stdev)
    plt.errorbar(it[i], time[i], yerr=error[i], marker="x")
    plt.title('Zależnośc czasu obliczania od maksymalnej liczby iteracji')
    plt.xlabel('Liczba iteracji')
    plt.ylabel('Czas obliczeń')
    plt.grid()
plt.show()

plt.scatter(it, newton_norm_ans2)
plt.grid()
plt.title('Rozwiązanie 2-go miejsca zerowego metodą Newtona')
plt.xlabel('Liczba iteracji')
plt.ylabel('Dokładność rozwiązania')
plt.semilogy()
plt.show()

time = []
error = []
for i in range(len(it)):
    method_time = %timeit -r 10 -n 10 -o main.newton(fun, dfun, ddfun, a2, b2, epsilon, it[i])
    time.append(method_time.average)
    error.append(method_time.stdev)
    plt.errorbar(it[i], time[i], yerr=error[i], marker="x")
    plt.title('Zależnośc czasu obliczania od maksymalnej liczby iteracji')
    plt.xlabel('Liczba iteracji')
    plt.ylabel('Czas obliczeń')
    plt.grid()
plt.show()
```
