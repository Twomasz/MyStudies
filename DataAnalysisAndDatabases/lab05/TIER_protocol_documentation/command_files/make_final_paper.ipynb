```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
top_10_divided = pd.read_csv('../analysis_data/top_10_countries_divided_by_types_of_alcohol.csv')
top_10_total = pd.read_csv('../analysis_data/top_10_countries_pure_alcohol.csv')
melted_top_5_total = pd.read_csv('../analysis_data/top_5_countries_all_alcohols.csv')
sum_all = pd.read_csv('../analysis_data/sum_of_different_alcohol_servings.csv')
```

```python
fig = plt.figure(figsize = (20, 5))
ax = fig.add_subplot(111)
tab = plt.table(cellText = top_10_divided.values[:, 1:],
         rowLabels = top_10_divided.index+1,
         colLabels = ['Country','Beer Servings','Country','Spirit Servings','Country','Wine Servings'],
         colWidths = [2]*7,
         bbox=[0, -1, 1, 2],
         )
tab.auto_set_font_size(False)
tab.set_fontsize(16)
ax.set_title("Who drinks the most beer, spirits and wine?", fontsize=30)
ax.axis("off")
plt.show()
```

```python
sns.barplot(data=top_10_total,
            x='total_litres_of_pure_alcohol',
            y='country',
            orient='h')
plt.title('Where people drink the most of pure alcohol?')
plt.xlim([9, 15])
plt.grid()
plt.show()
```

```python
sns.barplot(data=melted_top_5_total,
            x='country',
            y='quantity',
            hue='type_of_alcohol')
plt.xticks(rotation=45)
plt.grid()
plt.title('Which alcohols are drunk in 5 countries where people drink the most')
plt.show()
```

```python
plt.pie(sum_all.values.tolist()[0][1:],
        labels=sum_all.columns.tolist()[1:],
        autopct='%.0f%%',
        textprops={'size': 'larger'},
        startangle=90)
plt.title('Sum of drunk alcohol servings around the world')
plt.show()
```
